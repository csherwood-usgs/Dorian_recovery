```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
import numpy as np
import glob

ddir = r'D:/crs/proj/2023_satellite_imagery/SanJose/'
base = r'20170819_162151_1009'
fn = base + '_3B_AnalyticMS_SR_clip.tif'
metadata = base + '_metadata.json'
```

```python
flist = glob.glob(ddir+'*clip.tif')
print(flist)
```

```python
# Demo calcs with one file
for i, f in enumerate(flist):
    # band order is blue, green, red, nir
    ds = rasterio.open(flist[i]) 
    # no. of bands and image resolution
    print( ds.count, ds.height, ds.width)
    # Coordinate Reference System
    print( ds.crs )
    b = ds.read(1)
    g = ds.read(2)
    r = ds.read(3)
    n = ds.read(4)
    ndvi = (n.astype(float) - r.astype(float))/(n + r)
    ndwi = (g.astype(float) - n.astype(float))/(n + g)
    fig, (ax0, ax1)  = plt.subplots(1,2, sharey=True)
    p0=ax0.imshow(ndvi, cmap="RdYlGn")
    plt.colorbar(p0, ax=ax0)
    p1=ax1.imshow(ndwi, cmap="bwr_r")
    plt.colorbar(p1, ax=ax1)
    fig_name = flist[i].replace("_3B_AnalyticMS_SR_clip.tif", ".jpg")
    print(fig_name)
    plt.savefig(fig_name, dpi=200, bbox_inches='tight')

    # Set spatial characteristics of the output object to mirror the input
    kwargs = ds.meta
    kwargs.update(
        dtype=rasterio.float32,
        count = 1)

    ndvi_fn = flist[i].replace("_3B_AnalyticMS_SR_clip", "_ndvi")
    print(ndvi_fn)
    # todo - change this to write a COG
    with rasterio.open(ndvi_fn, 'w', **kwargs) as dst:
            dst.write_band(1, ndvi.astype(rasterio.float32))
    # Create the ndwi file
    ndwi_fn = flist[i].replace("_3B_AnalyticMS_SR_clip", "_ndwi")
    print(ndwi_fn)
    # todo - change this to write a COG
    with rasterio.open(ndwi_fn, 'w', **kwargs) as dst:
            dst.write_band(1, ndwi.astype(rasterio.float32))
```
